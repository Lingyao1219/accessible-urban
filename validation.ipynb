```python
import pandas as pd
import numpy as np
import krippendorff
```

### Cross validation

```python
cross_df1 = df1[['no','text','target_text','annotation']]
cross_df2 = df2[['no','text','target_text','annotation']]
cross_df = pd.merge(cross_df1, cross_df2, on=['no','text','target_text'], how='inner')
cross_df
```

```python
label_map = {'negative': -1, 
             'neutral': 0, 
             'positive': 1,
             'unrelated': 2}

cross_df['annotation_x'] = cross_df['annotation_x'].map(label_map)
cross_df['annotation_y'] = cross_df['annotation_y'].map(label_map)

reliability_data = np.array([cross_df['annotation_x'], cross_df['annotation_y']])
alpha = krippendorff.alpha(reliability_data=reliability_data, level_of_measurement='nominal')

print(f"Krippendorff's alpha: {alpha}")
```
